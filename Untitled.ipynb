### Imputing the age column for us as we want to fill the na values

#### Working on Age

```python
from sklearn.neighbors import KNeighborsRegressor
def knn_age_impute(df,target_feature='Age'):
    df_impute = df.copy()
    numeric_df = df_impute.select_dtypes(np.number)
    non_na_columns = numeric_df.loc[: ,numeric_df.isna().sum() == 0].columns
    #Above will remove age
    y_train = numeric_df.loc[numeric_df[target_feature].isna() == False, target_feature]
    X_train = numeric_df.loc[numeric_df[target_feature].isna() == False, non_na_columns]
    X_test = numeric_df.loc[numeric_df[target_feature].isna() == True, non_na_columns]
    knn = KNeighborsRegressor()
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    df_impute.loc[df[target_feature].isna() == True, target_feature] = y_pred
    return df_impute
```

```python
# Age before computation
sns.histplot(df['Age'], color= 'teal', kde= True);
```

```python
df_impute=knn_age_impute(df=df)
# Age after computation
sns.histplot(df_impute['Age'], color= 'teal', kde= True);
```

### Data TransFormation: Encoding And Scaling

```python
data=df_impute.copy()
```

```python
data.info()
```

```python
# converting dtype to object if not  manually
data['Pclass'] = data['Pclass'].apply(str)
target = target.apply(str)
```

```python
data.info()
```

```python
# we are using labelEncoder as we know
## we have ordinal data of sex and Embarked
data = pd.get_dummies(data, columns=["Sex"], prefix=["Sex"],drop_first=False)
data.head()
```

```python
le = LabelEncoder()
numerical_embarked = le.fit_transform(data['Embarked'])
data['Embarked'] = numerical_embarked
```

```python
# fially our data is :
data.head()
```

### Feature Engineering

#### Both the SibSp and Parch column suggests whether the person was person was travelling with his family or not. So we will convert these features into a single feature called family

```python
data['Family'] = np.where(data['SibSp'] + data['Parch'] > 0, 1, 0)
data.drop(['SibSp', 'Parch'], axis= 1, inplace= True)
```

```python
data.head()
```

### Scaling

```python
sc = StandardScaler()
data[['Age', 'Fare']] = sc.fit_transform(data[['Age', 'Fare']])
```

### Let's Apply Grid SearchCV

```python
model1=DecisionTreeClassifier()
```

```python
parameter ={
    "criterion" : ["gini","entropy"],
    "max_depth" : range(2,32,1),
    "max_features": ['sqrt', 'log2', None],
    "min_samples_leaf": range(1,25,1),
    "min_samples_split": range(2,25,1),
    "splitter" : ['best', 'random']
}
```

```python
grid_search = GridSearchCV(estimator=model1,
                           param_grid=parameter,
                           cv=5,
                           n_jobs =-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_parameters = grid_search.best_params_
print(best_parameters)
```

```python
clf = DecisionTreeClassifier(criterion = best_parameters['criterion'],
                             max_depth =best_parameters['max_depth'], 
                             max_features=best_parameters['max_features'],
                             min_samples_leaf= best_parameters['min_samples_leaf'], 
                             min_samples_split= best_parameters['min_samples_split'], 
                             splitter =best_parameters['splitter'])
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
print('Accuracy score for train data is:', accuracy_score(y_train,clf.predict(X_train)))
print('Accuracy score for test data is:', accuracy_score(y_test,clf.predict(X_test)))
```
